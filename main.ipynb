```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
dataset = pd.read_csv('dataset/new_dataset.csv')
dataset_num = dataset.select_dtypes([int,float])
scaler = StandardScaler()
dataset_scaled = scaler.fit_transform(dataset_num)
dataset_scaled = pd.DataFrame(data=dataset_scaled, columns=dataset_num.columns)
```

```python
dataset_scaled.to_csv('dataset_scaled.csv', index=False)
```

```python
dataset_scaled
```

```python
dataset['raw_score'] = (
    0.3 * dataset_scaled['repay_ratio'] +
    0.2 * dataset_scaled['redeem_ratio'] +
    0.2 * dataset_scaled['times_deposit'] +
    0.1 * dataset_scaled['times_redeemed'] +
    0.1 * dataset_scaled['times_repayed'] +
    -0.2 *dataset_scaled['liquidation_count']
)

dataset['credit_score'] = ((dataset['raw_score'] - dataset['raw_score'].min()) /
                      (dataset['raw_score'].max() - dataset['raw_score'].min()) * 1000).astype(int)

score_dict = pd.DataFrame(data = zip(dataset['wallet_address'], dataset['credit_score']), columns=['wallet_address','credit_score'])

score_dict.to_json('credit_scores.json')

print(dataset[['wallet_address', 'credit_score']].head())
```
